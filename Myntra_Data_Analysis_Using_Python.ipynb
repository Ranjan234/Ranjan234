```
# which month highest collected revenue.
orders['Month'] = orders['Date'].dt.strftime('%B')
```

```
orders
```

```
orders['Total Price'] = orders['Original Price'] - (orders['Original Price'] * orders['Discount%'])
```

```
#groupby funtion
gb = orders.groupby('Month')['Total Price'].sum().reset_index()
gb
```

```
# visualize data
plt.figure(figsize=(12, 4))
sns.barplot(x='Month',hue = gb.index, palette='GnBu', y='Total Price', data=gb)
plt.xticks(rotation=60)
plt.show()
```

```
# Jion two table
merged_df = pd.merge(orders, products, on='Product ID', how='inner')
merged_df
```
