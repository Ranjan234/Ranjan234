**View** **and** **copy**
-copy is new array, and the view is just a view of the original array.
-Any changes made to the copy will not affect original array.
- Any changes made to the view will affect the original array.

````python
import numpy as np
````

````python
arr=np.array([1,2,3,4])
````

````python
#copy method
copy=arr.copy()
copy[0]=42
print(arr)
print(copy)
````

````python
Q3
````

````python
IQR=Q3-Q1
````

````python
IQR
````

````python
low=Q1-1.5*IQR
````

````python
low
````

````python
upp=Q3+1.5*IQR
````

````python
upp
````

````python
low,upp
````

````python
Outliers=[]
````

````python
list_
````

````python
Outliers
````
