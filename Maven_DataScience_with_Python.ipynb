```python
groceries.head()
```

```python
groceries['Shipment_Date_DOW'] = groceries['Shipment_Date_DOW'].map(dow_mapping)
```

```python
groceries['New_Shipment_Date'] = groceries.Next_Scheduled_Shipment + pd.to_timedelta(1, 'd')
```

# Grouping Data
- The split-apply-combine approach is used to group data in a DataFrame and
apply calculations to each group
- You can group a DataFrame by using the .groupby() method.

```python
groceries.groupby('Category')['Inventory'].sum()
```

```python
groceries.groupby('Category')['Inventory'].count()
```

```python
groceries.groupby('Category')['Price_Dollars'].max()
```

```python
groceries.groupby('Category')[['Item','Price_Dollars']].max()
```

```python
(groceries[['Category', 'Item' ,'Price_Dollars']]
         .sort_values(by='Price_Dollars', ascending=False)
          .groupby('Category')
          .head(1))
```

# ASSIGNMENT: EXPLORING DATA
 1. Filter out any data before 2010 and after 2019
 2. Group the data by country and calculate the
maximum happiness score for each one
 3. Sort the grouped countries by happiness score
and return the top five
 4. Group the data by country and calculate the
average happiness score for each one
 5. Sort the grouped countries by happiness score
and return the top five
 6. Compare the two lists

```python
import pandas as pd
```

```python
df= pd.read_csv("/content/happiness_survey_data.csv")
```

```python
df.head()
```

```python
# There are 2089 rows and 6 columns.
df.shape
```

```python
# the years range from 2005 - 2021
# hapiness scores range from 2 - 8
df.describe()
```

```python
# Filter out any data before 2010 and after 2019.
mask = (df.year >= 2010) & (df.year <= 2020)
filtered_df = df[mask]
filtered_df.head()
```

```python
# Groupby  the data by country and calculate the maximum hapiness score for each one.
filtered_df.groupby('country_name')['happiness_score'].max()
```

```python
# Sort the grouped countries by happiness score and return the top five.
filtered_df.groupby('country_name')['happiness_score'].max().sort_values(ascending=False).head(5)
```

```python
# Group the  data by country and calculate the average happiness score for each one.
filtered_df.groupby('country_name')['happiness_score'].mean()
```

```python
# sort the grouped countries by happineaa score and return the top five
filtered_df.groupby('country_name')['happiness_score'].mean().sort_values(ascending=False).head(5)
```
