# creating New Column

```python
import pandas as pd
```

```python
groceries = pd.read_excel("/content/Groceries.xlsx")
```

```python
groceries.head(5)
```

```python
groceries.Price_Dollars * 1.05
```

```python
round(groceries.Price_Dollars * 1.05, 2)
```

```python
groceries["New_Column"] = round(groceries.Price_Dollars * 1.05, 2)
```

```python
groceries.head(5)
```

```python
groceries.Inventory.sum()
```

```python
groceries["Total Inventory"] = groceries.Inventory.sum()
```

```python
round(groceries["Inventory"] / groceries["Total Inventory"] * 100, 2)
```

```python
groceries["Inventory %"] = round(groceries["Inventory"] / groceries["Total Inventory"] * 100, 2)
```

```python
groceries.head(5)
```

```python
import numpy as np
```

```python
groceries["Low Inventory"] = np.where(groceries.Inventory < 50, 'Low Inventory', '')
```
